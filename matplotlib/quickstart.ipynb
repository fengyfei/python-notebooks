```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-whitegrid')
plt.rc('figure', figsize=(8, 6))
```

```python
x = np.linspace(-4 * np.pi, 4 * np.pi, 200)

plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
x = np.linspace(-4 * np.pi, 4 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y, 'o', color='blue')
plt.show()
```

```python
rng = np.random.RandomState(0)

for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(4), rng.rand(4), marker, label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8)
plt.show()
```

```python
x = np.linspace(-4 * np.pi, 4 * np.pi, 100)
y = np.sin(x)

plt.scatter(x, y, marker='o')
plt.show()
```

```python
rng = np.random.RandomState(0)

for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.scatter(rng.rand(4), rng.rand(4), marker=marker, label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(32)
y = rng.randn(32)

colors = rng.rand(32)
sizes = 512 * rng.rand(128)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='copper')
plt.colorbar()
```

```python
def f(x, y):
    return np.sin(x) + np.cos(y)

x = np.arange(-2.0 * np.pi, 2.0 * np.pi, 0.1)
y = np.arange(-2.0 * np.pi, 2.0 * np.pi, 0.1)

X, Y = np.meshgrid(x, y)
c = plt.contour(X, Y, f(X, Y), cmap='copper')
plt.clabel(c, inline=True, fontsize=8)
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```
