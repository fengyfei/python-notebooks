```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-colorblind')
plt.rc('figure', figsize=(8, 6))

x = np.linspace(0, 4 * np.pi, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend()
```
