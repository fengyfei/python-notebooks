```python
googc = googc.sort_values(by=['date'])
```

```python
googc
```

```python
axes = plt.subplot(1, 1, 1)
axes.set_xticks([])
axes.plot(googc['date'], googc['open'])
```

```python
googc['date']=pd.to_datetime(googc['date'],format='%Y-%m-%d')
googc.index = googc['date']

month = googc.groupby(pd.Grouper(freq='M'))

monthDF = month.mean()
monthDF
```

```python
axes = plt.subplot(1, 1, 1)
axes.set_xticks([])
axes.plot(monthDF.index, monthDF['open'])
```

```python
axes = plt.subplot(1, 1, 1)
axes.set_xticks([])
axes.scatter(monthDF.index, monthDF['open'])
```
